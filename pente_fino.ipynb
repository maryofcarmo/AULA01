# Leitura dos arquivos e colocar todos em um dicionário

```python
# Defina o diretório onde estão os arquivos CSV
diretorio = '/content/drive/MyDrive/TCC/DADOS/CONVENCIONAL'

# Obtenha a lista de todos os arquivos CSV no diretório
arquivos_csv = [f for f in os.listdir(diretorio) if f.endswith('.csv')]

arquivos_csv
```

```python
# dicionário para armazenar os DataFrames
dataframes = {}


# Loop para ler cada arquivo CSV e armazenar no dicionário
for arquivo in arquivos_csv:

    caminho_completo = os.path.join(diretorio, arquivo)

    #obtem o nome do arquivo sem a sua extensão e coloca na variável 'nome_df'
    nome_df = os.path.splitext(arquivo)[0]

    '''lê o dataframe pelo caminho do arquivo completo,delimitado por ponto e
    vírgula (CSV) e exclui da leitura as 10 primeiras linhas do arquivo
    com essa informação joga na variável 'dataframes[nome_df]'''
    dataframes[nome_df] = pd.read_csv(caminho_completo, delimiter=';', skiprows=10)
```

# Separa os df de acordo com o tipo de index

```python
# Função para verificar e classificar os índices dos DataFrames
def classificar_indices(dataframes):
    index_com_valores = []
    index_datatime = []
    index_numerico = []
    index_outro = []

    for nome, df in dataframes.items(): #nome: chave, df:valor
        if isinstance(df.index, pd.RangeIndex) and (df.index.start == 0 and df.index.stop == len(df) and df.index.step == 1):
            # Índice padrão do pandas (0, 1, 2, ...)
            index_numerico.append(nome)
        elif isinstance(df.index, pd.DatetimeIndex):
            # Índice do tipo datetime
            index_datatime.append(nome)
        elif not isinstance(df.index, pd.RangeIndex):
            # Índice não é numérico (não é o padrão do pandas)
            index_com_valores.append(nome)
        else:
            # Qualquer outro caso que não seja detectado acima
            index_outro.append(nome)

    return index_com_valores, index_numerico, index_outro, index_datatime
```

```python
# Chamar a função para classificar os índices
index_com_valores, index_numerico, index_datatime, index_outro = classificar_indices(dataframes)

# Exibir os resultados
print("DataFrames com índice preenchido com algum valor:", index_com_valores)
print("DataFrames com índice numérico padrão do pandas:", index_numerico)
print("DataFrames com índice que não se encaixam em nenhum dos padrões acima:", index_outro)
print("DataFrames com índice com valor datatime:", index_datatime)
```

```python
#contanto a quantidade de elemento de cada lista
print("Quantidade de DataFrames com índice preenchido com algum valor:" + str(len(index_com_valores)))
print("Quantidade de DataFrames om índice numérico padrão do pandas:" + str(len(index_numerico)))
```

# Transformação em série temporal

```python
#colunas que eu quero nas minhas séries
colunas_interesse = ['PRECIPITACAO TOTAL, MENSAL(mm)', 'INSOLACAO TOTAL, MENSAL(h)', 'VENTO, VELOCIDADE MEDIA MENSAL(m/s)']

# Dicionário para armazenar as séries temporais
series_temp = {}

# Loop sobre cada DataFrame no dicionário original
for nome, df in df_index_correto.items():
    # Selecionar e copiar as colunas de interesse
    df_copia = df[colunas_interesse].copy()

    # Converter as colunas para o tipo float
    for coluna in colunas_interesse:
        df_copia[coluna] = df_copia[coluna].astype(str).str.replace(',', '.')
        df_copia[coluna] = df_copia[coluna].astype(float)

    # Adicionar o DataFrame copiado ao dicionário de séries temporais
    series_temp[nome] = df_copia
```

# LIMPEZA

```python
############################### TESTE ##################################

# Inicializar o novo dicionário para armazenar as séries limpas
series_limp = {}

# Loop sobre cada série temporal no dicionário original
for nome, df in series_temp.items():
    # Substituir 'null' (ou 'None') por 'NaN'
    df = df.replace({None: np.nan})

    # Remover as linhas que têm qualquer valor de NaN
    df_limpo = df.dropna()

    # Adicionar o DataFrame limpo ao dicionário 'series_limp'
    series_limp[nome] = df_limpo

# Agora 'series_limp' contém as séries temporais limpas
```

```python
# Inicializar uma lista para armazenar o nome das séries que ainda contêm NaN
series_com_nan = []

# Iterar sobre cada série no dicionário 'series_limp'
for nome, df in series_limp.items():
    # Verificar se há algum NaN no DataFrame
    if df.isna().any().any():
        # Se houver NaN, adicionar o nome da série à lista
        series_com_nan.append(nome)

# Exibir a lista de séries que ainda contêm valores NaN
print("Séries com valores NaN restantes:", series_com_nan)
```

Função para limpar uma série, removendo linhas INTEIRAS caso ocorra penas um valor 'null' ou NaN em uma célula de uma coluna

```python
def limpar_serie(serie):
    # Substitui a string 'null' por NaN para uniformizar a limpeza
    serie.replace('null', np.nan, inplace=True)

    # Remove linhas que contenham qualquer NaN
    serie_limpa = serie.dropna()

    return serie_limpa
```

Aplicando a função de limpeza para cada série do dicionário de séries temporais 'series_temp', criando outro dicionário denominado 'series_temp_limp'

```python
# dicionário teste
SL = {}

for nome, serie in series_temp.items():
  SL[nome] = limpar_serie(serie)

# teste

SL['Nome:_BARREIRAS_Codigo_Estacao:_83236']
```
